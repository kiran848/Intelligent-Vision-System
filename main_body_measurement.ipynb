```python
import subprocess

# List of Python scripts to run
scripts = [
    "full_height.py",
    "shoulder.py",
    "arm_length.py",
    "chest.py",
    "waist.py",
    "lower_length.py"
]

# Start each script in sequence
outputs = []

for script in scripts:
    try:
        # Run each script and capture output
        process = subprocess.Popen(
            ["python", script], 
            stdout=subprocess.PIPE, 
            stderr=subprocess.PIPE
        )
        
        # Wait for the process to complete and capture output
        stdout, stderr = process.communicate()  # Capture output and errors
        outputs.append(stdout.decode())  # Decode byte output to string
        
        if stderr:
            print(f"Error in script {script}: {stderr.decode()}")
    
    except Exception as e:
        print(f"Error starting script {script}: {e}")

# Print the output of each script
for i, script in enumerate(scripts):
    print(f"Output from {scripts[i]}:\n{outputs[i]}")

print("All scripts have finished executing.")
```
