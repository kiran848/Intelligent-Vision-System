```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    "Full Body Length": [160, 170, 180, 175, 165],
    "Waist": [70, 80, 90, 85, 75],
    "Chest": [90, 100, 110, 105, 95],
    "Shoulder": [40, 45, 50, 47, 43],
    "Lower Length": [70, 75, 80, 77, 73],
}

df = pd.DataFrame(data)

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", cbar=True)
plt.title("Correlation Between Body Measurements")
plt.show()
```

```python
import numpy as np

# Sample predicted vs actual values
actual = [160, 170, 180, 175, 165]
predicted = [158, 172, 178, 176, 167]

# Calculate absolute errors
errors = np.abs(np.array(actual) - np.array(predicted))

# Heatmap for error visualization
error_data = pd.DataFrame([errors], columns=["Full Body Length", "Waist", "Chest", "Shoulder", "Lower Length"])

plt.figure(figsize=(8, 2))
sns.heatmap(error_data, annot=True, fmt=".1f", cmap="Reds", cbar=True, yticklabels=["Error"])
plt.title("Prediction Errors Across Measurements")
plt.show()
```

```python
# Normalize data for better visual impact
normalized_data = (df - df.min()) / (df.max() - df.min())

plt.figure(figsize=(8, 6))
sns.heatmap(normalized_data, annot=True, cmap="YlGnBu", cbar=True)
plt.title("Normalized Dataset Distribution of Body Measurements")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Example training data
X = df.drop(columns=["Full Body Length"])
y = df["Full Body Length"]

# Fit Random Forest model
model = RandomForestRegressor()
model.fit(X, y)

# Feature importance
feature_importance = model.feature_importances_

# Heatmap
importance_df = pd.DataFrame([feature_importance], columns=X.columns)

plt.figure(figsize=(8, 2))
sns.heatmap(importance_df, annot=True, cmap="Greens", cbar=True, yticklabels=["Importance"])
plt.title("Feature Importance for Full Body Length Prediction")
plt.show()
```
